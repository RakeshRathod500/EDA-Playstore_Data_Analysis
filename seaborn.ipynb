```python
df2=sns.load_dataset('fmri')
```

```python
sns.relplot(x='timepoint',y='signal',kind='line',data=df2)
```

```python
# shadow (blue)====> confidence interval
sns.relplot(x='timepoint',y='signal',kind='line',data=df2,ci='sd')
```

```python
# to remove the ci ===> estimator=none
sns.relplot(x='timepoint',y='signal',kind='line',data=df2,ci='sd',estimator=None)
```

```python
df2
```

```python
sns.relplot(x='timepoint',y='signal',kind='line',data=df2,ci='sd',style='region',hue='event')
```

```python
sns.relplot(x='timepoint',y='signal',kind='line',data=df2,ci='sd',style='region',hue='event',markers=True)
```
